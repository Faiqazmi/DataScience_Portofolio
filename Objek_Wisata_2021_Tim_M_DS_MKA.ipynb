## Wisata Alam Berdasarkan Jenisnya di Kalimantan TImur

```python
df_alam_percentage = df_alam
df_alam_percentage = df_alam_percentage.drop(['KABUPATEN / KOTA', 'NO', 'JUMLAH'], axis=1)
df_alam_percentage = df_alam_percentage.iloc[10]
df_alam_percentage = df_alam_percentage.to_list()
df_alam_percentage_value = df_alam_percentage
df_alam_percentage_columns = ['SUNGAI', 'DANAU PEDESAAN', 'TELUK', 'GUNUNG',
       'AIR TERJUN', 'AIR PANAS', 'WISATA GOA', 'TAMAN HUTAN RAYA',
       'TAMAN NASIONAL', 'TAMAN WISATA LAUT', 'HUTAN ADAT', 'HUTAN MANGROVE',
       'HUTAN LINDUNG']
df_alam_percentage_explode = []
for i in range(13):
  df_alam_percentage_explode.append(0.05)

df_alam_percentage_explode = tuple(df_alam_percentage_explode)
colors = ['#ff9999','#66b3ff', '#CE49BF', '#FFD93D', '#8479E1', '#6EDCD9', '#E45826', '#019267', '#D82148', '#5902EC', '#573391', '#BB6464', '#F76E11']


fig = plt.gcf()
fig.set_size_inches(8,8)
plt.pie(df_alam_percentage_value, colors = colors, labels=df_alam_percentage_columns, autopct='%1.1f%%', startangle=90, pctdistance=0.65, explode = df_alam_percentage_explode)
centre_circle = plt.Circle((0,0),0.30,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.tight_layout()
plt.title("Wisata Alam Yang Ada Di Kalimantan Timur")
plt.show()
```
