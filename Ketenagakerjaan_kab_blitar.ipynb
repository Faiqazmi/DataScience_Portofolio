```python
df_tenagaKerja.columns = ['provinsi', 'kabupaten_kota', 'kecamatan', 'status', 'kode_pum', 'desa_kelurahan', 'tanggal', 'kategori', 'laki-laki', 'perempuan', 'jumlah', 'rasio']
```

```python
def cek(x):
  if len(x.split()) == 2:
    l = list()
    for i in x.split():
      a = i.capitalize()
      l.append(a)
    hasil = ' '.join(l)
    return hasil
  else:
    kata = x.capitalize()
    return kata
```

```python
cl = ['provinsi', 'kabupaten_kota', 'kecamatan', 'status', 'desa_kelurahan']

for i in cl:
  df_tenagaKerja[i] = df_tenagaKerja[i].apply(cek)
```

```python
df_tenagaKerja.head(6)
```

```python
df_tenagaKerja['kategori'] = df_tenagaKerja['kategori'].apply(lambda x: x.lower())
```

```python
df_tenagaKerja.head(5)
```

```python
df_tenagaKerja.info()
```

```python
col = ['laki-laki',	'perempuan', 'jumlah']

for i in col:
  df_tenagaKerja[i] = df_tenagaKerja[i].astype(float).astype(int)
```

```python
df_tenagaKerja['rasio'] = df_tenagaKerja['rasio'].map(lambda x: str(x.replace(',','.')))
```

```python
df_tenagaKerja['rasio'] = df_tenagaKerja['rasio'].astype(float)
```

```python
df_tenagaKerja.info()
```

```python
df_tenagaKerja['tanggal'] = df_tenagaKerja['tanggal'].map(lambda x : datetime.strptime(x, '%d/%m/%Y'))
```

```python
df_tenagaKerja.info()
```

```python
df_tenagaKerja = df_tenagaKerja.reindex(columns=['tanggal', 
                                                 'provinsi',	
                                                 'kabupaten_kota',	
                                                 'kecamatan',	
                                                 'status', 
                                                 'desa_kelurahan', 
                                                 'kode_pum', 
                                                 'kategori',
                                                 'laki-laki',
                                                 'perempuan',	
                                                 'jumlah',	
                                                 'rasio'])
```

```python
#sheet1
df_tenagaKerja_original = df_tenagaKerja.copy()
df_tenagaKerja_original.index += 1
df_tenagaKerja_original
```

```python
#sheet2
df_tenagaKerja_sortByDate = df_tenagaKerja.copy()
df_tenagaKerja_sortByDate.sort_values(by='tanggal', inplace=True)
df_tenagaKerja_sortByDate.reset_index(drop=True, inplace=True)
df_tenagaKerja_sortByDate.index += 1
df_tenagaKerja_sortByDate
```

```python
df_tenagaKerja = df_tenagaKerja[df_tenagaKerja['tanggal'] > '2018-01-01'].reset_index(drop=True)
df_tenagaKerja
```

```python
#sheet3
df_kategori_per_kecamatan = df_tenagaKerja.pivot_table(index='kecamatan', columns='kategori', values='jumlah', aggfunc='sum')
df_kategori_per_kecamatan.fillna(0, inplace=True)

for i in df_kategori_per_kecamatan.columns:
  df_kategori_per_kecamatan[i] = df_kategori_per_kecamatan[i].astype(int)
df_kategori_per_kecamatan
```

```python
df_kategori_per_kecamatan.info()
```

```python
#sheet4
df_kategori_per_gender = df_tenagaKerja.pivot_table(index='kategori', values=['laki-laki', 'perempuan'], aggfunc='sum')
df_kategori_per_gender.reset_index(inplace=True)
df_kategori_per_gender.index +=1
df_kategori_per_gender['total'] = df_kategori_per_gender['laki-laki'] + df_kategori_per_gender['perempuan']
df_kategori_per_gender
```

```python
#sheet5
df_tenagaKerja_year = df_tenagaKerja.set_index('tanggal')
df_tenagaKerja_year = df_tenagaKerja_year.resample('Y').sum()
df_tenagaKerja_year.reset_index(inplace=True)
df_tenagaKerja_year.drop(columns=['jumlah', 'rasio'],inplace=True)
df_tenagaKerja_year.rename({'tanggal':'Tahun'}, axis=1 ,inplace=True)
df_tenagaKerja_year.index = [1,2,3,4] 
df_tenagaKerja_year
```

```python
with pd.ExcelWriter('Data_Ketenagakerjaan_Kab_Blitar.xlsx') as writer:
  df_tenagaKerja_original.to_excel(writer, sheet_name='Original')
  df_tenagaKerja_sortByDate.to_excel(writer, sheet_name='Original_byDate')
  df_kategori_per_kecamatan.to_excel(writer, sheet_name='Jumlah_Kategori_per_Kercamatan')
  df_kategori_per_gender.to_excel(writer, sheet_name='Jumlah_Kategori_per_Gender') 
  df_tenagaKerja_year.to_excel(writer, sheet_name='Total_Kategori_per_Tahun')
```
